## 1) Download USGS N10 Landslide Susceptibility

```python
# Download ScienceBase JSON metadata
url = "https://prod-is-usgs-sb-prod-publish.s3.amazonaws.com/65ccea5bd34ef4b119cb3bac/n10_susc.zip"
out_folder  = os.path.join(data_path, "usgs_n10_susc")
os.makedirs(out_folder, exist_ok=True)
out_filename = "n10_susc.zip"
zip_path = os.path.join(out_folder, out_filename)

# Download the file
if os.path.exists(zip_path):
    print(f"File already exists: {out_filename}")
else:
    response = requests.get(url, stream=True, timeout=120)
    response.raise_for_status()
    with open(zip_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=1024 * 1024):
            if chunk:
                f.write(chunk)
    print(f"Downloaded: {out_filename}")

# unzip and extract data. Check what is included in the zip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(out_folder)

# sanity check
print(f'Extracted files: {os.listdir(os.path.join(out_folder, "n10_susc"))}')
```
