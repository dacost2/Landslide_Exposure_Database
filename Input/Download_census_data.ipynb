## 5) Urban-Rural classification  
Data at the Census Block Level. Select year of Census Information for the analysis

```python
# Select year of information to download
year = 2025

# Base url for the download
url = f"https://www2.census.gov/geo/tiger/TIGER{year}/TABBLOCK20/"

# List of files to download for all 56 states/territories
data_files = [f"tl_{year}_{str(i).zfill(2)}_tabblock20.zip" for i in range(1, 57)]
# Create a downloads directory
out_folder  = os.path.join(data_path, "census_gov_data", "tabblock20")
os.makedirs(out_folder, exist_ok=True)

print(f"Downloading {len(data_files)} Census Blocks files...")
downloaded_files = []
for file_name in data_files:
    file_url = url + file_name
    file_path = os.path.join(out_folder, file_name)
    
    # Skip if already downloaded
    if os.path.exists(file_path):
        print(f"Already exists: {file_name}")
        downloaded_files.append(file_path)
        continue

    print(f"Downloading {file_name}")
    try:
        response = requests.get(file_url, stream=True, timeout=60)
        response.raise_for_status()
        
        # Write file in chunks to handle large downloads
        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)
        print(f"✓ Downloaded: {file_name}")
    except requests.exceptions.RequestException as e:
        print(f"✗ Error downloading {file_name}: {e}")

    # Load the shapefile for the current state and append to the main block GeoDataFrame
    if os.path.exists(file_path):
        block_gdf = gpd.read_file(os.path.join(out_folder, file_name))
    if i == file_name[0]:
        block = block_gdf.copy()
    else:
        block = pd.concat([block, block_gdf], ignore_index=True)

print(f"\nDownload and extraction complete! Downloaded {len(downloaded_files)}/{len(data_files)} files")

block.to_file(os.path.join(out_folder, f"tl_{2025}_all_states_tabblock20.gpkg"), driver="GPKG")
# sanity check
print(f'Block data has {len(block)} records and the following columns: \n{block.columns.tolist()}\n')
print(block.head(3))
```

## 7) ACS (2018-2022)

```python
# 5-year data. Select year of information to download
year = 2022

url = f"https://www2.census.gov/geo/tiger/TIGER_DP/{year}ACS/"

# List of ACS {year} 5-year GDB files to download (tract level for all 56 states/territories)
data_files = [f"ACS_{year}_5YR_TRACT_{str(i).zfill(2)}.gdb.zip" for i in range(1, 57)]

# Create a downloads directory
out_folder  = os.path.join(data_path, "census_gov_data", "ACS_2022_5YR_TRACT", "gdb_files")
os.makedirs(out_folder, exist_ok=True)

print(f"Downloading {len(data_files)} Census Blocks files...")
downloaded_files = []
for file_name in data_files:
    file_url = url + file_name
    file_path = os.path.join(out_folder, file_name)
    
    # Skip if already downloaded
    if os.path.exists(file_path):
        print(f"Already exists: {file_name}")
        downloaded_files.append(file_path)
        continue
    
    print(f"Downloading {file_name}...")
    try:
        response = requests.get(file_url, stream=True, timeout=60)
        response.raise_for_status()
        
        # Write file in chunks to handle large downloads
        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)
        print(f"✓ Downloaded: {file_name}")
        
        # Extract the zip file
        extract_dir = os.path.join(out_folder, file_name.replace('.zip', ''))
        with zipfile.ZipFile(file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_dir)
        print(f"✓ Extracted: {file_name}")
        downloaded_files.append(file_path)
        
    except requests.exceptions.RequestException as e:
        print(f"  ✗ Error downloading {file_name}: {e}")
    except Exception as e:
        print(f"  ✗ Error processing {file_name}: {e}")
print(f"\nDownload and extraction complete! Downloaded {len(downloaded_files)}/{len(data_files)} files")
```

```python
# Extract poverty and income tables from all GDB files
from pathlib import Path

output_dir = os.path.join(data_path, "census_gov_data", "ACS_2022_5YR_TRACT", "extracted_tables")
os.makedirs(output_dir, exist_ok=True)

# Initialize lists to collect data from all states
poverty_data_list = []
income_data_list = []

poverty_table = "X17_POVERTY"
income_table = "X19_INCOME"

print("Extracting poverty and income tables from GDB files...")

# Get all outer GDB directories (ACS_2022_5YR_TRACT_XX.gdb)
outer_gdb_dirs = sorted([d for d in Path(os.path.join(data_path, "census_gov_data", "ACS_2022_5YR_TRACT", "gdb_files")).iterdir() if d.is_dir() and d.name.startswith(f'ACS_{year}_5YR_TRACT_')])
print(f"Found {len(outer_gdb_dirs)} outer GDB directories\n")

for outer_gdb in outer_gdb_dirs:
    state_id = outer_gdb.name.split('_')[4].replace('.gdb', '')  # Extract state number (01-56)
    print(f"Processing {outer_gdb.name}...")
    
    # Find nested state GDBs inside the outer GDB
    nested_gdbs = sorted([d for d in outer_gdb.iterdir() if d.is_dir() and d.name.endswith('.gdb')])
    
    for nested_gdb in nested_gdbs:
        state_name = nested_gdb.name
        print(f" -> Reading nested GDB: {state_name}")
        
        try:
            # List all layers/tables in the nested GDB
            try:
                layers = gpd.list_layers(nested_gdb)
                layer_names = layers['name'].tolist() if isinstance(layers, pd.DataFrame) else layers
            except:
                # If list_layers fails, try to read and catch the error message
                print(f" ! Could not list layers, attempting direct reads...")
                layer_names = []
            
            # Check for poverty table (X17_POVERTY)
            poverty_found = False
            for layer_name in layer_names:
                if poverty_table in layer_name.upper():
                    try:
                        poverty_gdf = gpd.read_file(nested_gdb, layer=layer_name)
                        poverty_gdf['STATE_ID'] = state_id
                        poverty_gdf['STATE_NAME'] = state_name
                        poverty_data_list.append(poverty_gdf)
                        print(f" ✓ Extracted {layer_name}: {len(poverty_gdf)} records")
                        poverty_found = True
                        break
                    except Exception as e:
                        print(f" ✗ Error reading {layer_name}: {e}")
            
            # Check for income table (X19_INCOME)
            income_found = False
            for layer_name in layer_names:
                if income_table in layer_name.upper():
                    try:
                        income_gdf = gpd.read_file(nested_gdb, layer=layer_name)
                        income_gdf['STATE_ID'] = state_id
                        income_gdf['STATE_NAME'] = state_name
                        income_data_list.append(income_gdf)
                        print(f" ✓ Extracted {layer_name}: {len(income_gdf)} records")
                        income_found = True
                        break
                    except Exception as e:
                        print(f" ✗ Error reading {layer_name}: {e}")
            
            # If no tables found by name, print available layer names for debugging
            if not poverty_found and not income_found and layer_names:
                print(f" ! No poverty/income tables found. Available layers:")
                for ln in layer_names[:5]:
                    print(f"      - {ln}")
        
        except Exception as e:
            print(f" ✗ Error processing {state_name}: {e}")

print(f"\n\nCombining tables across all states...")

# Combine all poverty data
if poverty_data_list:
    poverty_combined = pd.concat(poverty_data_list, ignore_index=True)
    print(f"✓ Combined poverty data: {len(poverty_combined)} total records from {len(poverty_data_list)} states")
    print(f"  Columns: {poverty_combined.columns.tolist()}")
    
    # Save to csv
    poverty_csv = os.path.join(output_dir, "ACS_2022_X17_POVERTY_combined.csv")
    poverty_combined.to_csv(poverty_csv, index=False)
    print(f" ✓ Saved to: {poverty_csv}")
else:
    print("✗ No poverty data was extracted")
    poverty_combined = None

# Combine all income data
if income_data_list:
    income_combined = pd.concat(income_data_list, ignore_index=True)
    print(f"\n✓ Combined income data: {len(income_combined)} total records from {len(income_data_list)} states")
    print(f"  Columns: {income_combined.columns.tolist()}")
    
    # Save to csv
    income_csv = os.path.join(output_dir, "ACS_2022_X19_INCOME_combined.csv")
    income_combined.to_csv(income_csv, index=False)
    print(f"  ✓ Saved to: {income_csv}")
else:
    print("✗ No income data was extracted")
    income_combined = None

print(f"\n✓ Extraction complete!")
```
